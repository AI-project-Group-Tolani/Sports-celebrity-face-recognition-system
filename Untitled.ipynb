```python
import cv2
import matplotlib.pyplot as plt

def cropped_image(image_path, x_start, y_start, x_end, y_end):
    # Read the original image
    original_image = cv2.imread(image_path)
    
    if original_image is None:
        print("Error: Unable to read the image.")
        return
    
    # Crop the image using the provided parameters
    cropped = original_image[y_start:y_end, x_start:x_end]
    
    # Plot the original and cropped images
    plt.figure(figsize=(10, 5))
    
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    
    plt.subplot(1, 2, 2)
    plt.title("Cropped Image")
    plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
    
    plt.tight_layout()
    plt.show()

# Define the image path and cropping parameters
image_path = r'D:\Sumit\SEM_5\AI\Project\AI-project\test_images\Maria sharapova\1_5e56899d423a0.jpg'
x_start, y_start, x_end, y_end = 100, 100, 300, 300

# Call the function
cropped_image(image_path, x_start, y_start, x_end, y_end)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

def crop_eyes(image_path):
    # Load the image
    original_image = cv2.imread(image_path)
    
    # Load a pre-trained face and eye detection classifier
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
    
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
    
    # Detect faces in the image
    faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    
    # Iterate over detected faces
    for (x, y, w, h) in faces:
        face_roi = gray_image[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(face_roi)
        
        # Assume there are two eyes detected
        if len(eyes) >= 2:
            cropped_eyes = []
            for (ex, ey, ew, eh) in eyes:
                cropped_eye = face_roi[ey:ey+eh, ex:ex+ew]
                cropped_eyes.append(cropped_eye)
            
            return cropped_eyes

    return None

# Path to the image
image_path = r'D:\Sumit\SEM_5\AI\Project\AI-project\test_images\Maria sharapova\1_5e56899d423a0.jpg'

# Call the function to get cropped eyes
cropped_eyes = crop_eyes(image_path)

if cropped_eyes:
    # Display the original image
    original_image = cv2.imread(image_path)
    plt.figure(figsize=(8, 4))
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
    plt.title("Original Image")
    
    # Display the cropped left eye
    plt.subplot(1, 3, 2)
    plt.imshow(cropped_eyes[0], cmap='gray')
    plt.title("Left Eye")
    
    # Display the cropped right eye
    plt.subplot(1, 3, 3)
    plt.imshow(cropped_eyes[1], cmap='gray')
    plt.title("Right Eye")
    
    plt.tight_layout()
    plt.show()
else:
    print("Unable to detect two eyes.")
```

```python
import numpy as np 
import pywt
import cv2

def w2d(img, mode='haar' ,  level=1):
    imArray = img
    #Datatype conversions
    #Convert to grayscale
    imArray = cv2.cvtColor(imArray,cv2.COLOR_RGB2GRAY)
    #Convert to float
    imArray = np.float32(imArray)
    imArray /= 255;
    #computer cofficients
    coeffs = pywt.wavedec2(imArray, mode, level=level)
    
    #Process Coefficients
    coeffs_H = list (coeffs)
    coeffs_H[0] *= 0;
    
    #reconstruction
    imArray_H = pywt.waverec2(coeffs_H, mode);
    imArray_H *= 255;
    imArray_H = np.uint8(imArray_H)
    
    return imArray_H 
    
```

```python
class_dict={}
count =0
for celebrity_name in celebrity_file_name_dict.keys():
    class_dict[celebrity_name] = count
    count = count +1
class_dict
```

```python
x, y = [], []


for celebrity_name, training_files in celebrity_file_name_dict.items():
    for training_image in training_files:
        img = cv2.imread(training_image)
        if img is None:
            continue
        scalled_raw_img = cv2.resize(img, (32, 32))
        img_har = w2d(img,'db1',5)
        scalled_img_har = cv2.resize(img_har,(32, 32))
        combined_img = np.vstack((scalled_raw_img.reshape(32*32*3,1),scalled_img_har.reshape(32*32,1)))
        x.append(combined_img)
        y.append(celebrity_name)
```

# Data cleaning process is done. Now we are ready to train our Model
We will use SVM with rbf kernel tuned with heuristic finetuning

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y, random_state=0)
pipe = Pipeline([('scaler', StandardScaler()),('svc', SVC(kernel = 'rbf', C = 10))])
pipe.fit(x_train, y_train)
pipe.score(x_test, y_test)
```

```python
print(classification_report(y_test,  pipe.predict(x_test)))
```

# Let's use GridSearch to try out different paramets.Goal is to come up with best modle with   best fine tuned parameters

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
model_params ={
        'svm' : {
                'model': svm.SVC (gamma= 'auto', probability=True),
                'params' : {
                        'svc__C': [1, 10, 100, 1000],
                        'svc__kernel': ['rbf', 'linear']
                }
        },
        'random_forest' : {
                'model' : RandomForestClassifier(),
                'params' : {
                        'randomforestclassifier__n_estimators': [1,5,10]
                }
        },
        'logistic regression' : {
                'model': LogisticRegression (solver='liblinear', multi_class='auto'),
                'params': {
                        'logisticregression__C' :  [1,5,10]
                }
        }
}
```

```python
scores = []
best_estimators ={}
import pandas as pd
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp['model'])
    clf = GridSearchCV(pipe, mp['params'], cv=6, return_train_score =False)
    clf.fit(x_train, y_train)
    scores.append({
        'model' : algo,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df
```
